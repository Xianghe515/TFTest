```python
# 한글 형태소 분석을 위한 오픈 소스 한글 형태소 분석(Twitter(Okt))
twitter = Okt()

result = []       # 형태소 분석 결과를 담을 리스트
for line in rdw:     # 5개의 데이터만 사용
    malist = twitter.pos(line[1], norm=True, stem=True)     # 형태소 분석
    r = []
    for word in malist:
        if not word[1] in ['Josa', 'Eomi', 'Punctuation']:  # 조사, 어미, 문장부호 제외
            r.append(word[0])
    rl = (" ".join(r)).strip()  # 형태소 사이에 (" ")을 넣고, 양쪽 공백을 제거
    result.append(rl)
```

# 불용어 제거       
*불용어: 문장 내에서 빈번하게 발생하여 의미를 부여하기 어려운 단어
- 'a', 'the' 같은 단어들은 모든 구문에서 빈번하게 사용 -> 특정 의미 파악하기 어려움
- 불용어는 자연어 처리에 있어 효율성을 떨어뜨리고, 처리 시간은 늘리는 단점이 있음
- 불용어 제거는 자연어 처리의 성능을 향상시키는 방법 중 하나

```python
# 불용어 처리 예제(NLTK 사용)
import nltk
from nltk.corpus import stopwords       # 불용어 말뭉치     *corpus : 말뭉치
# nltk.download('stopwords')
# nltk.download('punkt')
from nltk.tokenize import word_tokenize

# sample
sample_text = "One of the first things that we ask ourselves is what are the pros and cons of any task we perform."
# tokenize
text_tokens = word_tokenize(sample_text)

# 불용어 처리
tokens_without_sw = [word for word in text_tokens if not word in stopwords.words('english')]
print("불용어 제거 미적용 : ", text_tokens, "\n")
print("불용어 제거 적용 : ", tokens_without_sw, "\n")
```

# 어간 추출(Stemming), 표제어 추출(Lemmatization)
● 어간 추출(Stemming) : 단어의 어간을 추출하는 과정
- 예 : Automate, Automatic, Automation -> Automat

● 표제어 추출(Lemmatization) : 단어의 원형을 추출하는 과정
- 예 : am, are, is -> be

● 어간 추출과 표제어 추출은 둘 다 어근 추출이 목적

※ 차이점 : 어간 추출은 사전에 없는 단어도 추출할 수 있지만, 표제어 추출은 사전에 있는 단어만 추출할 수 있음

```python
# NTLK를 이용한 불용어 처리 예제

# 포터 알고리즘 : 단어의 원형이 비교적 보존되어 있음
print("포터 알고리즘")
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

print(stemmer.stem('obesses'), stemmer.stem('obssesed'))
print(stemmer.stem('standardizes'), stemmer.stem('standardization'))
print(stemmer.stem('national'), stemmer.stem('nation'))
print(stemmer.stem('absentness'), stemmer.stem('absently'))
print(stemmer.stem('tribalical'), stemmer.stem('tribalicalized'), "\n")

# 랭커스터 알고리즘 : 단어의 원형을 더 많이 축소 -> 정밀도가 떨어짐
print("랭커스터 알고리즘")
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()

print(stemmer.stem('obesses'), stemmer.stem('obssesed'))
print(stemmer.stem('standardizes'), stemmer.stem('standardization'))
print(stemmer.stem('national'), stemmer.stem('nation'))
print(stemmer.stem('absentness'), stemmer.stem('absently'))
print(stemmer.stem('tribalical'), stemmer.stem('tribalicalized'))
```

```python
# 표제어 추출
# 일반적으로 어간 추출보다 표제어 추출의 성능이 더 좋음
# 품사, 문법, 뜻 등을 고려하여 단어의 원형을 추출
# 단점 : 처리 속도가 어간 추출보다 느림

import nltk
# nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize('obsesses'), lemmatizer.lemmatize('obsessed'))
print(lemmatizer.lemmatize('standardizes'), lemmatizer.lemmatize('standardization'))
print(lemmatizer.lemmatize('national'), lemmatizer.lemmatize('nation'))
print(lemmatizer.lemmatize('absentness'), lemmatizer.lemmatize('absently'))
print(lemmatizer.lemmatize('tribalical'), lemmatizer.lemmatize('tribalicalized'))
```

```python
# 예제로 사용할 데이터 - covtype.csv    *https://github.com/jeffheaton/data/blob/master/covtype.csv
# 데이터 설명 : https://archive.ics.uci.edu/ml/datasets/Covertype
```

```python
# 정규화를 위한 라이브러리
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.python.data import Dataset
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import models
from tensorflow.keras import layers, Input

# 데이터 읽어오기
df = pd.read_csv('./data/covtype.csv')
# df.info()
# df.isnull().sum()

# 데이터 분리
x = df[df.columns[:54]]
y = df['cover_type']
```

```python
# 훈련 데이터와 검증 데이터 생성 - sklearn의 train_test_split() 사용
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=90)      # 훈련 데이터 70%, 검증 데이터 30%
```

```python
# 모델 생성
model = tf.keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(8, activation='softmax')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=26,
    batch_size=60
)
```

```python
# 모델의 정확도 시각화
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager

# matplotlib에는 한글 인코딩 값이 없음 -> 한글 깨짐 현상
font_fname = 'C:/Windows/Fonts/malgun.ttf'      # 주피터 노트북에서 한글 깨지는 문제 해결
font_family = font_manager.FontProperties(fname=font_fname).get_name()      # 폰트 패밀리 값 알아오기
plt.rcParams["font.family"] = font_family       # plt에 폰트 설정

# 모델 정확도 그래프 그리기
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(accuracy))

plt.plot(epochs, accuracy, label="훈련 데이터셋")
plt.plot(epochs, val_accuracy, label="검증 데이터셋")
plt.legend()
plt.title('정확도')
plt.figure()

plt.plot(epochs, loss, label="훈련 데이터셋")
plt.plot(epochs, val_loss, label="검증 데이터셋")
plt.legend()
plt.title('오차')
plt.show()
```
